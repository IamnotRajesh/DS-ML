SIMPLE LINEAR REGRESION

- In this lab we will use simple regression to predict car price.
- Objective:
    - to know if the dealer is offering the fair value for my trade-in.
    - To know if I put a fair value on my car.

DATASET

- We will use the dataset of Car price prediction hosted on IBM Cloud object.

IMPORT PACKAGES


```python
import os #  for file operations
import numpy as np #  for numerical processing
import pandas as pd #   data frames
import matplotlib.pyplot as plt #  plotting library
import seaborn as sns #   statistical data visualization
from sklearn.model_selection import train_test_split # modeling
from sklearn.linear_model import LinearRegression # modeling
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score #modeling
```

```python
Data_path = '/'.join(os.path.abspath('').split('/')[:-4])+'automobileEDA.csv'
df = pd.read_csv(Data_path)
df.head()
```

```python
# get features
x = df.drop('price', axis=1) # independent features
y = df['price'] # dependent features
```

TRAIN TEST SPLIT

```python
# random_state = shuffles from the fixed position

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

SIMPLE LINEAR REGRESSION

Predict car price using "highway.mpg"

```python
# get simple linear independent variables
x_train_simple = x_train['highway-mpg'].values.reshape(-1, 1)
x_test_simple = x_test['highway-mpg'].values.reshape(-1, 1)
```

```python
x_train_simple
```

```python
x_test_simple
```

```python
# Fit simple linear regression model
simple_lm =LinearRegression()
simple_lm.fit(x_train_simple, y_train)
```

```python
# get prediction
y_hat_simple = simple_lm.predict(x_test_simple)
y_hat_simple[:5]
```

```python
# get intercept
simple_lm.intercept_
```

```python
# get slope
simple_lm.coef_.item()
```

ESTIMATED AMOUNT

    - Price = 39260.28546717645(simple_lm.intercept_) - 845.0027947574558(simple_lm.coef_.item()) * highway-mpg
