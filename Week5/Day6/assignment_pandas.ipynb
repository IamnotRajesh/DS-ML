<li>Read 'car_details.csv' file and create a pandas dataframe from this file.</li>
<li>Find the maximum price for each of the car brand.</li>
<li>Find the average price for each of the fuel types.</li>
<li>Find the average km_driven for each of the seller_types.</li>
<li>Find the count of each of the car names.</li>
<li>Find the maximum km_driven for each of the owner types.</li>

```python
import pandas as pd
```

```python
df = pd.read_csv('car_details.csv')
df.head()
```

```python
# Maximum price of each car brand

max_price = df.groupby('name')['selling_price'].max().reset_index(name= 'max_price')
print(max_price)
```

```python
# Average price for each fuel type
avg_price = df.groupby('fuel')['selling_price'].mean().reset_index(name= 'average_price')
avg_price
```

```python
#avg Km driven 
avg_km = df.groupby('seller_type')['km_driven'].mean().reset_index(name= 'average_km')
print (avg_km)
```

```python
# count of each of the car names
count_name = df.groupby('name')['name'].count().reset_index(name='count')
print(count_name)
```

```python
# maximum km driven by owner
max_km = df.groupby('owner')['km_driven'].max().reset_index(name= 'max_km')
print(max_km)
```

#### Question
<li>Read 'Fortune_1000.csv' using pandas read_csv() method.</li>
<li>Find the company headquartered in Los Angeles with the largest number of employees.</li>
<li>Select only the rows that have a city name equal to Los Angeles.</li>
<li>Use DataFrame.sort_values() to sort those rows by the employees column in descending order.</li>
<li>Use DataFrame.iloc[] to select the first row from the sorted dataframe.</li>


```python
# Read 'Fortune_1000.csv' using pandas read_csv() method:
import pandas as pd

df = pd.read_csv('Fortune_1000.csv')
df.head()
```

```python
# Select only the rows that have a city name equal to Los Angeles:
la_df = df[df['city'] == 'Los Angeles']
la_df
```

```python
# Use DataFrame.sort_values() to sort those rows by the employees column in descending order:
sorted_df = la_df.sort_values('num. of employees', ascending=False)
sorted_df
```

```python
# Use DataFrame.iloc[] to select the first row from the sorted dataframe:
first_row = sorted_df.iloc[0]
first_row
```

#### Question
<li>Read 'car_details.csv' file and create a pandas dataframe from this file.</li>
<li>Find the maximum price for each of the car brand.</li>
<li>Find the average price for each of the fuel types.</li>
<li>Find the average km_driven for each of the seller_types.</li>
<li>Find the count of each of the car names.</li>
<li>Find the maximum km_driven for each of the owner types.</li>

```python
import pandas as pd

# Read 'car_details.csv' file and create a pandas dataframe
df = pd.read_csv('car_details.csv')
df.head()
```

```python
# Find the maximum price for each of the car brand
max_price_by_brand = df.groupby('name')['selling_price'].max().reset_index()

# Find the average price for each of the fuel types
avg_price_by_fuel_type = df.groupby('fuel')['selling_price'].mean().reset_index()

# Find the average km_driven for each of the seller_types
avg_km_driven_by_seller_type = df.groupby('seller_type')['km_driven'].mean().reset_index()

# Find the count of each of the car names
count_of_car_names = df['name'].value_counts().reset_index()

# Find the maximum km_driven for each of the owner types
max_km_driven_by_owner_type = df.groupby('owner')['km_driven'].max().reset_index()

print(max_price_by_brand)
print(avg_price_by_fuel_type)
print(avg_km_driven_by_seller_type)
print(count_of_car_names)
print(max_km_driven_by_owner_type)
```
