iloc

ways to slice pandas dataframe via integer value.

It allows you to select dataframe using integer indices for both rows and columns.

1st row is 0 and 2nd row is 1, and so on.

1st column is 0 and 2nd column is 1, and so on.

syntax:  df[row_index,column_index]

```python
df_hr.iloc[0,0]
```

```python
df_hr.iloc[:3, -3:]
```

loc

loc is used for label-based indexing.

allows us to slice dataframe using row label and column label instead of integer value as in loc.

syntax: df.loc[row_label, column_label]

```python
df_hr.loc[0, 'satisfaction_level']
```

```python
row_label = [0, 1, 2]
col_label = ['promotion_last_5years', 'Department', 'salary']

df_hr.loc[row_label, col_label]
```

Data Manipulation

Methods are:

replace()
- replaces value from dataframe
- df.replace({'old_val':'new_val', ...})

astype()
- change  data type of columns in a dataframe
- df.astype({'col':'new_type',...})

drop()
- drop column
- df.drop(col_name, axis=1)

sort_values()
- sort values either in ascending order or descending order
- df,sort_values(by=['column'],ascending=[True], inplace=[True])
- 

```python
df_hr['left'].replace({1: 'yes', 0: 'No'}, inplace=True)
df_hr['left']
```

```python
df_typecast = df_hr.astype({'left':'bool'})
df_typecast.head()
```
